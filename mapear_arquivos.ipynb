```python
import os
import pandas as pd

# --- CONFIGURAÇÃO ---
# Caminho exato que você me passou
caminho_da_pasta = "/Users/tomasdefreitascarvalho/Desktop/Work Space/Rubens/synapsee"

dados_arquivos = []

print(f"Lendo arquivos em: {caminho_da_pasta}...")

# Verifica se a pasta existe antes de começar
if not os.path.exists(caminho_da_pasta):
    print("ERRO: A pasta não foi encontrada. Verifique se o caminho está correto.")
else:
    # Percorre todas as pastas e subpastas
    for raiz, pastas, arquivos in os.walk(caminho_da_pasta):
        for arquivo in arquivos:
            # Ignora arquivos ocultos do Mac (aqueles que começam com .)
            if arquivo.startswith('.'):
                continue
                
            caminho_completo = os.path.join(raiz, arquivo)
            
            # Pega a extensão (ex: .ipynb, .csv)
            nome, extensao = os.path.splitext(arquivo)
            
            # Pega a subpasta (para saber se é Tunad, GAIA, etc)
            pasta_pai = os.path.basename(raiz)
            
            # Tenta pegar o tamanho do arquivo (em MB)
            try:
                tamanho_mb = os.path.getsize(caminho_completo) / (1024 * 1024)
            except:
                tamanho_mb = 0

            dados_arquivos.append({
                "Pasta Pai": pasta_pai,
                "Nome do Arquivo": arquivo,
                "Tipo": extensao.lower(),
                "Tamanho (MB)": round(tamanho_mb, 2),
                "Caminho Completo": caminho_completo
            })

    # Cria a tabela e salva
    if dados_arquivos:
        df = pd.DataFrame(dados_arquivos)
        nome_saida = "tabela_arquivos_synapsee.xlsx"
        df.to_excel(nome_saida, index=False)
        
        print("-" * 30)
        print(f"SUCESSO! Encontrei {len(df)} arquivos.")
        print(f"O arquivo Excel foi salvo como: {nome_saida}")
        print("-" * 30)
        
        # Mostra as primeiras 5 linhas para você conferir
        display(df.head()) # 'display' funciona melhor no Jupyter que 'print'
    else:
        print("Nenhum arquivo encontrado nessa pasta.")
```
